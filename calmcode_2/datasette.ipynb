```python
# now it's how to deploy datasette on cloud run
# --service=my-bigmac is to give it a name

datasette publish cloudrun bigmac.db --service=my-bigmac

# you will a get a url, to share it publicly
```

```python
# there are also other cloud providers

datasette publish heroku bigmac.db

or pip install datasette-publish-vercel
   datasette publish vercel bigmac.db --project my-vercel-project
    
or pip install datasette-publish-fly
   datasette publish fly bigmac.db --app="bigmac-app"
```

```python
# now it's about authentification. Make your document a private datafile.

# install datasette auth

pip install datasette-auth-passwords

# then you can hash a password. Get the hashed password in the 'metadata.json' file
```

```python
# run datasette with a password configured

datasette bigmac.db -m metadata.json
```

```python
# also deploy it on cloud run

datasette publish cloudrun bigmac.db --service=my-bigmac --install=datasette-vega -m metadata.json
```
